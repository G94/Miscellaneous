```python
import re
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv("segmentacion.txt", header = None)
```

```python
df.columns =["id_cliente"]
df.head()
```

```python
pattern_upper = re.compile("[A-Z]")
pattern_low = re.compile("[a-z]")
pattern_impar = re.compile("[13579]")
pattern_repetitive = re.compile("([a-zA-Z])\\1+")
pattern_vocal = re.compile("[aAeEiIoOuU]")
pattern_sequence = re.compile("jn|cg|ar|mp|fs|ic")

def condition_A(x):
    return (pattern_repetitive.search(x) and (len(pattern_vocal.findall(x))>=1))
   
def condition_B(x):
    return ((len(pattern_upper.findall(x)) >  len(pattern_low.findall(x))) or (len(pattern_impar.findall(x))>0))

def condition_C(x):
    return pattern_sequence.search(x)

def clasificacion(x):
    if condition_A(x): return "A"
    if condition_B(x): return "B"
    if condition_C(x): return "C"
    return "D"
```

```python
condition_B("xqPuDnYiNuX10v7okag1h2mayGnbw7")
```

```python
condition_A("Wzxepmkx1szRfsrctGinAqo6hgns9")
```

```python
if condition_C("gjDdjyslysFlqsolvglxipiKWztui"):
    print("sd")
```

```python
df["tipo_cliente"] = df.id_cliente.apply(clasificacion)
```
